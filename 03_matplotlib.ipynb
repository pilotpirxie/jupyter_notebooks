```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'green')
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('x lab')
axes.set_ylabel('y lab')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
for axis in axes:
    axis.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
fig.savefig('data/testfig.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Title')

ax.plot(x, x**2, label='X2')
ax.plot(x, x**3, label='X3')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green', linewidth=5, alpha=0.2, linestyle='dotted', marker='o', markersize=25, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='blue')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
```
